```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
l = [12,33,23,21,34,32]
l2 = [(12,33),(33,32),(23,21),(21,34),(34,32),(32,12),(12,23),(23,12)]
G.add_nodes_from(l)
G.add_edges_from(l2)
```

```python
nx.draw(G,with_labels=True)
plt.show()
```
