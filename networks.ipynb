```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
```

```python
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,4)
G.add_edge(4,1)
```

```python
nx.draw(G,with_labels=True)
plt.show()
```
